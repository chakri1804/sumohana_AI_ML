```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def Mean_centering(X):
    X = X.astype('float')
    for i in range(len(X)):
        X[i] = X[i] - np.mean(X[i])
    print(np.mean(X[0]))
    return X
```

```python
pic = Image.open('wildlife-bears2.jpg')
img = (np.array(pic))
```

```python
a,b,c = img.shape
print(a,b,c)
```

```python
Flat = img.reshape((a*b,c))
Flat = Flat.astype('float32')
```

```python
Flat = Flat.T
Flat.shape
```

```python
Flat = Mean_centering(Flat)
```

```python
Cx = (np.matmul(Flat,Flat.T)/(len(Flat[0])))
```

```python
Lx,Ex = np.linalg.eig(Cx)
```

```python
Y = np.matmul(Ex.T,Flat)
Y = Y.T
Y = Y.reshape(img.shape)
```

```python
Y = Y.astype(int)
```
