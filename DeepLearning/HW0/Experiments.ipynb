```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('cat.jpg')
plt.imshow(img)
```

```python
h,w,d = np.shape(img)
```

```python
b = np.zeros((h+20,w+20,d))
```

```python
plt.imshow(b)
```

```python
for i in range(d):
    b[:,:,i] += np.pad(img[:,:,i], ((10,10),(10,10)), 'constant')
b = b.astype(int)
plt.imshow(b)
```
